## 2. Implement Basic Logistic Regression

To implement basic logistic regression, we need to define the sigmoid function, cost function, and gradient descent algorithm. The sigmoid function maps any real-valued number into the (0, 1) interval, which is useful for binary classification:

Here is the function that we going to use for the sigmoid:

$$ \sigma(z) = \frac{1}{1 + e^{-z}} $$

Cost function:

$$ J(\theta) = -\frac{1}{m} \sum_{i=1}^{m} \left[ y^{(i)} \log(h_\theta(x^{(i)})) + (1 - y^{(i)}) \log(1 - h_\theta(x^{(i)})) \right] $$

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def compute_cost(X, y, w, b):
    m = len(y)
    z = np.dot(X, w) + b
    A = sigmoid(z)

    epsilon = 1e-8
    cost = (-1/m) * np.sum(y*np.log(A + epsilon) + (1-y)*np.log(1-A + epsilon))
    return cost
```

Now we define the gradients and the gradient descent algorithm to optimize the weights and bias:

$$ \text{Gradient}_w = \frac{1}{m} X^T (h_\theta(X) - y) $$

$$ \text{Gradient}_b = \frac{1}{m} \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)}) $$

The gradient descent algorithm iteratively updates the weights and bias using the computed gradients:
$$ w := w - \alpha \cdot \text{Gradient}_w $$
$$ b := b - \alpha \cdot \text{Gradient}_b $$

```python
def compute_gradients(X, y, w, b):
    m = len(y)
    z = np.dot(X, w) + b
    A = sigmoid(z)

    dw = (1/m) * np.dot(X.T, (A - y))
    db = (1/m) * np.sum(A - y)

    return dw, db
```

```python
def gradient_descent(X, y, w, b, alpha, iterations):
    costs = []

    for i in range(iterations):
        dw, db = compute_gradients(X, y, w, b)
        w -= alpha * dw
        b -= alpha * db

        if i % 100 == 0:
            cost = compute_cost(X, y, w, b)
            costs.append(cost)

    return w, b, costs
```

## Prediction

For the prediction we first define a function to predict probabilities that in this case is this equation:
$$ \hat{y} = \sigma(Xw + b) $$
and then we define a function to predict class labels based on a threshold (0.5):
$$ \text{predicted\_class} = \begin{cases} 1 &
\text{if } \hat{y} \geq 0.5 \\ 0 & \text{otherwise} \end{cases} $$


```python
def predict(X, w, b):
    probs = sigmoid(np.dot(X, w) + b)
    return (probs >= 0.5).astype(int)
```

## Evaluation metrics (brief)

Here we use common evaluation metrics for binary classification:
- Accuracy: proportion of correct predictions (TP+TN) / total.

$$ \text{Accuracy} = \frac{TP + TN}{TP + TN + FP + FN} $$

- Precision: TP / (TP + FP)  fraction of predicted positives that are true.

$$ \text{Precision} = \frac{TP}{TP + FP} $$

- Recall: TP / (TP + FN)  fraction of actual positives that are detected.

$$ \text{Recall} = \frac{TP}{TP + FN} $$

- F1 score: harmonic mean of precision and recall: 2 * (P * R) / (P + R).

$$ \text{F1 Score} = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}} $$

To avoid division by zero, we add a small epsilon to denominators in precision, recall, and F1 score calculations:

```python
def accuracy(y_true, y_pred):
    return np.mean(y_true == y_pred)

def precision(y_true, y_pred):
    tp = np.sum((y_true == 1) & (y_pred == 1))
    fp = np.sum((y_true == 0) & (y_pred == 1))
    return tp / (tp + fp + 1e-8)

def recall(y_true, y_pred):
    tp = np.sum((y_true == 1) & (y_pred == 1))
    fn = np.sum((y_true == 1) & (y_pred == 0))
    return tp / (tp + fn + 1e-8)

def f1_score(y_true, y_pred):
    p = precision(y_true, y_pred)
    r = recall(y_true, y_pred)
    return 2 * (p * r) / (p + r + 1e-8)
```

Now we can calculate these metrics for both training and testing sets to evaluate model performance and the metricts will be printed below:

```python
y_train_pred = predict(X_train, w, b)
y_test_pred = predict(X_test, w, b)

metrics = pd.DataFrame({
    "Metric": ["Accuracy", "Precision", "Recall", "F1 Score"],
    "Train": [
        accuracy(y_train, y_train_pred),
        precision(y_train, y_train_pred),
        recall(y_train, y_train_pred),
        f1_score(y_train, y_train_pred)
    ],
    "Test": [
        accuracy(y_test, y_test_pred),
        precision(y_test, y_test_pred),
        recall(y_test, y_test_pred),
        f1_score(y_test, y_test_pred)
    ]
})

metrics
```

## 3. Visualize Decision Boundaries

We can visualize the decision boundaries of the logistic regression model for different pairs of features. This helps us understand how the model separates classes based on feature values. We will create contour plots to show the decision boundary along with the training data points starting with training the model like this:

```python
def train_2feature_model(feature1, feature2):
    X_pair = df[[feature1, feature2]].values
    y_pair = df["Heart Disease"].values.reshape(-1,1)

    mu = X_pair.mean(axis=0)
    sigma = X_pair.std(axis=0)
    X_pair = (X_pair - mu) / sigma

    np.random.seed(42)
    m = len(y_pair)
    idx = np.random.permutation(m)
    train_size = int(0.7*m)
    train_idx = idx[:train_size]
    test_idx = idx[train_size:]

    X_train_pair = X_pair[train_idx]
    y_train_pair = y_pair[train_idx]

    w = np.zeros((2,1))
    b = 0

    w, b, _ = gradient_descent(X_train_pair, y_train_pair, w, b, alpha=0.01, iterations=2000)

    return w, b, mu, sigma
```

After that we can plot the decision boundary for each pair of features using contour plots. This visualizes how the logistic regression model separates the two classes based on the selected features with each feature od the dataset:

```python
def plot_decision_boundary(feature1, feature2):
    w, b, mu, sigma = train_2feature_model(feature1, feature2)

    X_pair = df[[feature1, feature2]].values
    y_pair = df["Heart Disease"].values

    X_pair = (X_pair - mu) / sigma

    x_vals = np.linspace(X_pair[:,0].min(), X_pair[:,0].max(), 100)
    y_vals = -(w[0]*x_vals + b) / w[1]

    plt.scatter(X_pair[:,0], X_pair[:,1], c=y_pair, cmap="bwr", alpha=0.7)
    plt.plot(x_vals, y_vals, color="black")
    plt.xlabel(feature1)
    plt.ylabel(feature2)
    plt.title(f"Decision Boundary: {feature1} vs {feature2}")
    plt.show()
```

## 4. Repeat with Regularization
Now we can repeat the training, prediction, evaluation, and visualization steps using now L2 regularization this time. Here are the functions with L2 regularization incorporated into the cost function and gradient descent:
- Cost function with L2 regularization:
$$ J(\theta) = -\frac{1}{m} \sum_{i=1}^{m} \left[ y^{(i)} \log(h_\theta(x^{(i)})) + (1 - y^{(i)}) \log(1 - h_\theta(x^{(i)})) \right] + \frac{\lambda}{2m} \sum_{j=1}^{n} w_j^2 $$
- Gradient descent with L2 regularization:
$$ \text{Gradient}_w = \frac{1}{m} X^T (h_\theta(X) - y) + \frac{\lambda}{m} w $$
$$ \text{Gradient}_b = \frac{1}{m} \sum_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)}) $$


```python
def compute_cost_reg(X, y, w, b, lambd):
    m = len(y)
    z = np.dot(X, w) + b
    A = sigmoid(z)

    epsilon = 1e-8
    cost = (-1/m) * np.sum(y*np.log(A + epsilon) + (1-y)*np.log(1-A + epsilon))

    reg_cost = (lambd / (2*m)) * np.sum(w**2)
    return cost + reg_cost
```

```python
def compute_gradients_reg(X, y, w, b, lambd):
    m = len(y)
    z = np.dot(X, w) + b
    A = sigmoid(z)

    dw = (1/m) * np.dot(X.T, (A - y)) + (lambd/m)*w
    db = (1/m) * np.sum(A - y)

    return dw, db
```

```python
def gradient_descent_reg(X, y, w, b, alpha, iterations, lambd):
    costs = []

    for i in range(iterations):
        dw, db = compute_gradients_reg(X, y, w, b, lambd)
        w -= alpha * dw
        b -= alpha * db

        if i % 100 == 0:
            cost = compute_cost_reg(X, y, w, b, lambd)
            costs.append(cost)

    return w, b, costs
```

```python
lambdas = [0, 0.001, 0.01, 0.1, 1]
results = []

for lambd in lambdas:
    w = np.zeros((X_train.shape[1],1))
    b = 0

    w_reg, b_reg, costs_reg = gradient_descent_reg(
        X_train, y_train, w, b,
        alpha=0.01, iterations=2000, lambd=lambd
    )

    y_pred_test = predict(X_test, w_reg, b_reg)

    results.append([
        lambd,
        accuracy(y_test, y_pred_test),
        precision(y_test, y_pred_test),
        recall(y_test, y_pred_test),
        f1_score(y_test, y_pred_test),
        np.linalg.norm(w_reg)
    ])
```

Once we have all the functions created we can see the data frame with the regularized model and then we can compare the performance metrics (accuracy, precision, recall, F1 score, w):

```python
reg_metrics = pd.DataFrame(results, columns=[
    "Lambda", "Accuracy", "Precision", "Recall", "F1 Score", "||w||"
])

reg_metrics
```

In this example we can see the cost comparation between the regularized and unregularized model being the regularized model the one that use lamda=0.1, this meaning that the regularized model has a slightly higher cost, indicating that the regularization term is adding to the overall cost but this can vary depending on the dataset and the value of lambda used and you can chaged it if you want:

```python
w0 = np.zeros((X_train.shape[1],1))
b0 = 0
w0, b0, costs0 = gradient_descent(X_train, y_train, w0, b0, 0.01, 2000)

w1 = np.zeros((X_train.shape[1],1))
b1 = 0
w1, b1, costs1 = gradient_descent_reg(X_train, y_train, w1, b1, 0.01, 2000, 0.1)

plt.plot(costs0, label="No Regularization")
plt.plot(costs1, label="L2 Regularization (=0.1)")
plt.title("Cost Comparison")
plt.xlabel("Iterations (per 100)")
plt.ylabel("Cost")
plt.legend()
plt.show()
```

Now we train the model with different values of lambda (e.g., 0.01, 0.1, 1, 10) and plot the cost curves for each to see how regularization affects convergence:

```python
def train_2feature_model_reg(feature1, feature2, lambd):
    X_pair = df[[feature1, feature2]].values
    y_pair = df["Heart Disease"].values.reshape(-1,1)

    mu = X_pair.mean(axis=0)
    sigma = X_pair.std(axis=0)
    X_pair = (X_pair - mu) / sigma

    np.random.seed(42)
    m = len(y_pair)
    idx = np.random.permutation(m)
    train_size = int(0.7*m)
    train_idx = idx[:train_size]

    X_train_pair = X_pair[train_idx]
    y_train_pair = y_pair[train_idx]

    w = np.zeros((2,1))
    b = 0

    w, b, _ = gradient_descent_reg(X_train_pair, y_train_pair, w, b, 0.01, 2000, lambd)
    return w, b, mu, sigma
```
